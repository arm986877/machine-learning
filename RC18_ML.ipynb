```python
import cv2 as cv
from glob import glob
```

```python
#define the directory path
folder_name = 'loop command'
path = '/content/drive/My Drive/' + folder_name
format = '.jpg'
# read all the images and convert them from BGR to RGB
images = [cv.imread(file) for file in glob(path + '/*' + format)]
#specify number of images
num_of_data = 4
#set up figure for plotting with specigic size and resolution
plt.figure(figsize=(32,2), dpi=150)
#loop through images and display them in a subplot
for i in range(4):
  plt.subplot(1, num_of_data, i+1)
  plt.title(i)
  plt.axis('off')
  plt.imshow(images[i])

```
